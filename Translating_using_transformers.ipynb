```python
from transformers import pipeline
```

```python
pipe = pipeline('text2text-generation', model= 'facebook/m2m100_418M')
```

```python
pipe('what is your hobby?',forced_bos_token_id=pipe.tokenizer.get_lang_id('ta'))
```
