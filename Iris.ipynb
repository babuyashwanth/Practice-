```python
import pandas as pd
import numpy as np
import pandas_bokeh
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
df.head()
```

```python
df['target'].unique()
```

```python
#plotting using andrews curve
```

```python
plt.figure(figsize=(12,10))
pd.plotting.andrews_curves(df,'target',colormap='Oranges')
plt.show()
```
